# Removing Null Values

```python
housing.isnull().values.any()
```

```python
#there are mainly 3 strategies to deal with nan values
#1.remove all the rows or columns containing nan values using dropna.
#2.remove the a specific row or column containing nan values using drop.
#3.replace nan values by using other values like mean median mode etc.
#this can be done using imputer or using fillna method. 
```

```python
#here i am using imputer
import numpy as np
from sklearn.impute import SimpleImputer
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')

```

```python
housing_nanremoved=imp_median.fit_transform(housing)
housing_nanremoved=pd.DataFrame(housing_nanremoved,columns=housing.columns)
housing_nanremoved.describe()
```

# test and train split using stratify train test split and feature scaling

```python
x=housing_nanremoved.drop('MEDV',axis=1)
from sklearn.preprocessing import StandardScaler
model_scaler=StandardScaler()
z=model_scaler.fit_transform(x)
x_nan_scaled=pd.DataFrame(z,columns=housing.columns.drop('MEDV'))

y=housing_nanremoved['MEDV'].copy()

print(x_nan_scaled)
print(y)


```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,stratify=x['CHAS'],test_size=0.25)
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

```

```python
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_predict=rfr.predict(x_test)
```

```python
mse=mean_squared_error(y_test, y_predict)
from math import sqrt
rmse=sqrt(mse)
print(mse)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(rfr,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
rmse_score=np.sqrt(-scores)
print(rmse_score)
```

```python
print(f"mean of scores : {rmse_score.mean()} and standard deviation of scores : {rmse_score.std()}")
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn import datasets
```

```python
digits=datasets.load_digits()
```

```python
digits.keys()
```

```python
data=pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
data
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
model_param={'svm':{'model':svm.SVC(gamma='auto'),'param':{'C':[1,5,10],'kernel':['rbf','linear', 'poly','sigmoid']}},
'RandomForestClassifier':{'model':RandomForestClassifier(max_depth=2, random_state=0),'param':{'n_estimators':[10,20,100]}},
'naive_bayes\(GuassianNB\)':{'model':GaussianNB(),'param':{ 'priors':[None]}},
'naive_bayes\(MultinomialNB\)':{'model':MultinomialNB(),'param':{'alpha':[1.0]}},
'DecisionTreeClassifier':{'model':DecisionTreeClassifier(),'param':{'criterion':['gini', 'entropy']}}} 
```

```python
from sklearn.model_selection import GridSearchCV
scores=[]
for model_name,model_prop in model_param.items():
    model=GridSearchCV(model_prop['model'],model_prop['param'],cv=5)
    model.fit(digits.data,digits.target)
    scores.append({'model':model_name,'model_params':model.best_params_,'model_score':model.best_score_})
scores
```

```python
conclusion=pd.DataFrame(scores)
```

```python
conclusion
```
